```python
#remove outliers

def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.15,0.6])
    IQR = Q1-Q3
    lower_range = Q1 - (0.75 * IQR)
    upper_range = Q3 + (0.75 * IQR)
    return lower_range,upper_range
```

```python
lowincome,uppincome=remove_outlier(DataAppend1['jumlah'])
DataAppend1['jumlah']=np.where(DataAppend1['jumlah']>uppincome,uppincome,DataAppend1['jumlah'])
DataAppend1['jumlah']=np.where(DataAppend1['jumlah']>lowincome,lowincome,DataAppend1['jumlah'])
DataAppend1.boxplot(column=['jumlah'])
plt.show()
```

```python
#logistik regeression
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X = DataAppend1[['tahun','bulan', 'kriteria_usaha','jumlah']]
y = DataAppend1['kelompok_bank']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50, random_state = 0)
```

```python
logistic_regression = LogisticRegression(solver='liblinear')
logistic_regression.fit(X_train, y_train)
y_pred = logistic_regression.predict(X_test)
```

```python
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

```python
print('Accuracy: ', metrics.accuracy_score(y_test, y_pred))
plt.show()
```
